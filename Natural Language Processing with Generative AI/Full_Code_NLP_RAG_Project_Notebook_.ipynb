```python
llama_llm = Llama(
    model_path=llama_model_path,
    n_ctx=5000,
    n_gpu_layers=38,
    n_batch=512
)
```

#### Response

```python
def response(query,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    model_output = mistral_llm(
      prompt=query,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    return model_output['choices'][0]['text']
```

### System and User Prompt Template

```python
qna_system_message = """
System Persona: You are the Lead Clinical Assistant. You serve as a bridge between high-volume medical
research and the clinical needs of healthcare professionals.
Your primary task is to prioritize answering what is asked in the question with clear, concise, accurate
and relevant information that can help healthcare professionals
make informed decisions. These questions can help answer about topics like Diagnostic Assistance,
Drug Information ,Treatment Plans, Specialty Knowledge, Critical Care Protocols etc.

Response Constraints (Strict Adherence Required):
1. Word Count: Every response must be between upto 300 words. If too long, consolidate lists and remove irrelevant text.
2. Structure: Use a multi-paragraph format. Separate distinct sections. Use numbered lists to ensure good readability.
3. Completeness: You must never provide a partial answer. Even if the topic is broad, summarize the answer to provide a finished clinical picture.

Mandatory Tone: Professional, intellectually honest, and clinical. Avoid conversational filler. Start directly with the information requested.
"""

qna_user_message_template = """
Context:
{context}

Question:
{question}
"""
```

### Response Function

```python
def generate_rag_response(user_input,chunk_size=500,k=3,max_tokens=500,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    if chunk_size == 500:
        retriever = retriever_500
    elif chunk_size == 1000:
        retriever = retriever_1000
    elif chunk_size == 2000:
        retriever = retriever_2000
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = mistral_llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message = """
You are a task-aware AI system responsible for evaluating the groundedness of an answer generated by an AI agent based on a provided context. Your goal is to rate how accurately the answer reflects the information in the context without introducing outside information or "hallucinations."

Evaluation Criteria
Rate the answer on a scale of 1 to 5 based on the following definitions:
* 1 (Not Grounded): The answer is entirely unsupported by the context or contradicts the context completely.
* 2 (Poorly Grounded): The answer contains significant factual errors or "hallucinations" relative to the context, with only minor or incidental alignment.
* 3 (Partially Grounded): The answer contains some accurate information from the context but also includes claims that are not supported or are extrapolated beyond what is written.
* 4 (Mostly Grounded): The answer is almost entirely supported by the context, though it may contain slight minor embellishments or a minor omission of nuance.
* 5 (Appropriately Grounded): The answer is perfectly supported by the context. Every claim made can be directly traced back to the provided text.

Response Format: Please output the evaluation in the following format
Score: [1-5]
Reasoning: [Provide a brief explanation justifying the score based on specific evidence or discrepancies found.]
"""
```

```python
relevance_rater_system_message = """
You are a task-aware AI system responsible for evaluating the relevance of an answer generated by an AI agent. Your goal is to determine how well the answer addresses the specific intent and requirements of the user's question within the given context.

Evaluation Criteria
Rate the relevance on a scale of 1 to 5 based on the following definitions:
* 1 (Irrelevant): The answer completely misses the point, ignores the question, or discusses a totally unrelated topic.
* 2 (Loosely Related): The answer mentions keywords from the question but fails to address the actual query or provides "filler" content.
* 3 (Partially Relevant): The answer addresses part of the user's question but ignores key constraints or leaves a significant portion of the query unanswered.
* 4 (Mostly Relevant):  The answer addresses the main intent of the question well, though it may include some slightly off-topic information or miss a minor nuance.
* 5 (Highly Relevant): The answer is perfectly aligned with the user's intent. It is concise, direct, and fully addresses all aspects of the question.

Response Format: Please output the evaluation in the following format
Score: [1-5]
Reasoning: [Provide a brief explanation justifying the score based on specific evidence or discrepancies found.]
"""
```

```python
user_message_template = """
Context:
{context}

Question:
{question}

Answer:
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input,k=3,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever_500.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = mistral_llm(
            prompt=prompt,
            max_tokens=800,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llama_llm(
            prompt=groundedness_prompt,
            max_tokens=200,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llama_llm(
            prompt=relevance_prompt,
            max_tokens=200,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```
